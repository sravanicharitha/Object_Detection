```python
import cv2 # pip install  opencv -python
```

```python
import matplotlib.pyplot as plt # pip install matplotlib
```

```python
config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
forzen_model = 'frozen_inference_graph.pb'
```

```python
model = cv2.dnn_DetectionModel(forzen_model,config_file)
```

```python
ClassLabels = []
file_name = 'coco.names'
with open(file_name, 'rt') as fpt:
    ClassLabels = fpt.read().rstrip('\n').split('\n')
```

```python
print(ClassLabels)
```

```python
print(len(ClassLabels))
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean((127.5,127.5,127.5))
model.setInputSwapRB(True)
```

# Reading Image

```python
img = cv2.imread('person.jpeg')
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
ClassIndex, confidece , bbox=model.detect(img,confThreshold=0.5)
```

```python
print(ClassIndex)
```

```python
print(confidece)
```

```python
font_scale=3
font= cv2.FONT_HERSHEY_PLAIN
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
    cv2.rectangle(img, boxes, (255,0,0), 2)
    cv2.putText(img,ClassLabels[ClassInd-1], (boxes[0]+10, boxes[1]+40), font, fontScale=font_scale, color=(0,255,0), thickness=3)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

# Video

```python
cap = cv2.VideoCapture("vediosample.mp4")
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Video can't open")
    
font_scale=3
font= cv2.FONT_HERSHEY_PLAIN
while True:
    ret,frame= cap.read()
    resize = cv2.resize(frame, (700, 500))
    
    ClassIndex, confidece, bbox=model.detect(resize,confThreshold=0.55)
    
    print(ClassIndex)
    
    if(len(ClassIndex)!=0):
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(),confidece.flatten(), bbox):
            if(ClassInd<=80):
                cv2.rectangle(resize,boxes,(255,0,0),2)
                cv2.putText(resize,ClassLabels[ClassInd-1],(boxes[0]+10, boxes[1]+40), font, fontScale=font_scale, color=(0,255,0), thickness=3)
    cv2.imshow('object detection Tutorial',resize)
    
    if cv2.waitKey(2)& 0xFF==ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
